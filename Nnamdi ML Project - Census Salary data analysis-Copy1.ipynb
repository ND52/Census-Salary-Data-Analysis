```python
#Importing Machine learning Models
from sklearn import metrics
from sklearn.model_selection import train_test_split
```

```python
#splitting the data into features and labels
X = Salaries2.drop(["salary", "final_weight", "occupation", "native-country", "relationship"], axis= "columns") # droping the label variable (salary) and other features that are not needed

y = Salaries2["salary"]
```

```python
#saving the result of one hot encoding
X = pd.get_dummies(X)
```

# Model Building 

Lets use the training data to build the model and then use test data to make prediction and evaluations.

we will compare the following  Machine Learning classificatoin algorithms and see how to carry out each of them, and their performance in comparison with each other:-

. Logistic Regression
. Random Forest Model
. K-Nearest Neighbors
. DecisionTreeClassifier
. GradientBoostingClassifier
. Naive Bayes Model
. Extreme Gradient Boost (XGBoost) Model

```python
#Splitting the data into training and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state= 1234)
```

# Logistic Regression 

```python
#Importing logistic regression from Sklearn module
from sklearn.linear_model import LogisticRegression
```

```python
#training the model
l_reg = LogisticRegression()

l_reg.fit(X_train, y_train)
```

```python
#prediction using the feature from test data
l_reg.predict(X_test)
```

```python
#saving the prediction
logistic_prediction = l_reg.predict(X_test)
```

```python
#preparing a confusion matrix to evaluate classification algorithms
def ConfusionMatrix(ytest, ypred, label = ["Negative", "Positive"]):
    
    from sklearn.metrics import confusion_matrix
   
    
    cm = confusion_matrix(ytest, ypred)
    
    plt.figure(figsize=(7, 5))
    
    sns.heatmap(cm, annot = True, cbar = False, fmt = 'd', cmap = 'Greens_r')
    
    plt.xlabel('Predicted', fontsize = 13)
    plt.xticks([0.5, 1.5], label)

    plt.yticks([0.5, 1.5], label)    
    plt.ylabel('Truth', fontsize = 13)
    
    plt.title('A confusion matrix');
```

```python
ConfusionMatrix(y_test, logistic_prediction, label= ["<=50k", ">50k"])
```

.There are 4594 True Negatives (TN): predicting that an individual will not make over 50k a year and truly an individual did not make over 50k a year

.There are 800 True positives (TP): predicting that an individual will  make over 50k a year and an individual actually made over 50k a year.

.There are 747 False Negatives (FN): predicting that an individual will not make over 50k a year and an individual actually made over 50k a year. 

.There are 367 False Positives (FP): predicting that an individual will make over 50k a year and an individual did not make over 50k a year.

```python
#Evaluation Metrics


#Checking the accuracy
Accuracy=metrics.accuracy_score(y_test, logistic_prediction)
Accuracy
```

The accuracy of the model is  82.88%

```python
#checking the F1 score
f1score=metrics.f1_score(y_test, logistic_prediction)
f1score
```

The F1 score of the model is  58.95%
